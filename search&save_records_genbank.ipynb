```python


handle = Entrez.esearch(db='nucleotide', term=search_info, retmax=len_rec)

record = Entrez.read(handle)

print(len(record["IdList"]))
id_list=record["IdList"]
hdl = Entrez.efetch(db='nucleotide', id=id_list, rettype='gb')
recs = list(SeqIO.parse(hdl, 'gb'))


base_path_save='/media/simon/data_files/genetic_data/coronaviridae/spikes_sequences/'
base_path_save2=base_path_save+'individual_seq/'

data=[]

for i, rec in enumerate(recs):
    print(rec.id)
    for feat in rec.features:
        #print(feat)
        
        if feat.type=='source':
            
            if 'isolation_source' in feat.qualifiers.keys():
                source=feat.qualifiers['isolation_source'][0]
            else:
                source='NaN'
            
            contry=feat.qualifiers['country'][0]
            date=feat.qualifiers['collection_date'][0]
        
        if feat.type=='CDS':
    
            if feat.qualifiers['gene'][0]=='S':
                
                #print(feat)
                #print('XX')
                #print(dir(feat))
                print(feat.qualifiers['protein_id'][0])
                #print(feat.location)
                a=feat.location
                #print(feat.qualifiers['translation'][0])
                #print(' ')
                
                prot_id=feat.qualifiers['protein_id'][0]
                prot_se=feat.qualifiers['translation'][0]
                rna_seq=str(rec.seq)[int(a.start):int(a.end)]
                isolate_id=rec.id
                
                
    rec1 = SeqRecord(Seq(prot_se), id=prot_id, name=isolate_id, description="S protein")           
    rec2 = SeqRecord(Seq(rna_seq), id=prot_id, name=isolate_id, description="S protein")
    
    
    rna_path=base_path_save2+'rna/seq_'+str(i)+'.fasta'
    prot_path=base_path_save2+'prot/seq_'+str(i)+'.fasta'
    SeqIO.write(rec2, rna_path, "fasta")
    SeqIO.write(rec1, prot_path, "fasta")
    
    
    data.append([isolate_id, contry, date, source, prot_id, rna_path, prot_path])
                
```
