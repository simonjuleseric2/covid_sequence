```python
from Bio import SeqIO
```

## combine lst of sequences to ref seq:
<br>
<br>

```python

lst_rec=[]
for record in SeqIO.parse("test_seq/ref_sequence.fasta", "fasta"):
    print(record.id)
    lst_rec.append(record)
    
for record in SeqIO.parse("test_seq/sequences.fasta", "fasta"):
    print(record.id)
    lst_rec.append(record)

SeqIO.write(lst_rec, "test_seq/all.fasta", "fasta")
```

```python
for record in SeqIO.parse("test_seq/all.fasta", "fasta"):
    print(record.id)
```
